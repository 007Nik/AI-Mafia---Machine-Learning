```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dft = pd.read_csv('Linear_X_Train.csv')
dft1 = pd.read_csv('Linear_Y_Train.csv')
test = pd.read_csv('Linear_X_Test.csv')
dfx = dft.values
dfy = dft1.values
tes = test.values
x = dfx.reshape(-1,1)
y = dfy.reshape(-1,1)
t = tes.reshape(-1,1)
X = (x-x.mean())/x.std()
Y = y
plt.plot(X,Y)
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(X,Y)
```

```python
prediction = Lr.predict(X)
Lr.score(X, Y)
```

```python
plt.scatter(X,Y,label = 'data')
plt.plot(X,prediction,color='red',label = 'prediction')
```
